### Importação dos dados



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_1.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science/refs/heads/main/base-de-dados-challenge-1/loja_4.csv"

loja = pd.read_csv(url)
loja2 = pd.read_csv(url2)
loja3 = pd.read_csv(url3)
loja4 = pd.read_csv(url4)

loja.head()
```

#1. Análise do faturamento


```python
# Cálculo do faturamento de cada loja
faturamento_loja1 = loja['Preço'].sum()
faturamento_loja2 = loja2['Preço'].sum()
faturamento_loja3 = loja3['Preço'].sum()
faturamento_loja4 = loja4['Preço'].sum()

    # Impressão dos resultados
print(f"Faturamento da Loja 1: R$ {faturamento_loja1:,.2f}")
print(f"Faturamento da Loja 2: R$ {faturamento_loja2:,.2f}")
print(f"Faturamento da Loja 3: R$ {faturamento_loja3:,.2f}")
print(f"Faturamento da Loja 4: R$ {faturamento_loja4:,.2f}")

```

# 2. Vendas por Categoria

# 3. Média de Avaliação das Lojas

```python
print("--- Média de Avaliações por Loja ---")

# Calcula a média para a Loja 1
media_avaliacoes_loja1 = loja['Avaliação da compra'].mean()
print(f"Média de Avaliações - Loja 1: {media_avaliacoes_loja1:.2f}")

# Calcula a média para a Loja 2
media_avaliacoes_loja2 = loja2['Avaliação da compra'].mean()
print(f"Média de Avaliações - Loja 2: {media_avaliacoes_loja2:.2f}")

# Calcula a média para a Loja 3
media_avaliacoes_loja3 = loja3['Avaliação da compra'].mean()
print(f"Média de Avaliações - Loja 3: {media_avaliacoes_loja3:.2f}")

# Calcula a média para a Loja 4
media_avaliacoes_loja4 = loja4['Avaliação da compra'].mean()
print(f"Média de Avaliações - Loja 4: {media_avaliacoes_loja4:.2f}")
```

# 4. Produtos Mais e Menos Vendidos

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Configurações para os gráficos
sns.set_style("whitegrid")
plt.rc('figure', figsize=(12, 8))

# --- Loja 1 ---
print("Análise de Vendas - Loja 1")
print("="*50)
vendas_produtos_loja1 = loja['Produto'].value_counts()
produto_mais_vendido_l1 = vendas_produtos_loja1.index[0]
produto_menos_vendido_l1 = vendas_produtos_loja1.index[-1]
print(f"Produto Mais Vendido: '{produto_mais_vendido_l1}' com {vendas_produtos_loja1.iloc[0]} unidades.")
print(f"Produto Menos Vendido: '{produto_menos_vendido_l1}' com {vendas_produtos_loja1.iloc[-1]} unidades.")

plt.figure()
top_10_loja1 = vendas_produtos_loja1.head(10)
sns.barplot(x=top_10_loja1.values, y=top_10_loja1.index, palette='viridis', hue=top_10_loja1.index, legend=False)
plt.title('Top 10 Produtos Mais Vendidos - Loja 1', fontsize=16)
plt.xlabel('Quantidade Vendida', fontsize=12)
plt.ylabel('Produto', fontsize=12)
plt.tight_layout()
plt.show()
print("\n\n")


# --- Loja 2 ---
print("Análise de Vendas - Loja 2")
print("="*50)
vendas_produtos_loja2 = loja2['Produto'].value_counts()
produto_mais_vendido_l2 = vendas_produtos_loja2.index[0]
produto_menos_vendido_l2 = vendas_produtos_loja2.index[-1]
print(f"Produto Mais Vendido: '{produto_mais_vendido_l2}' com {vendas_produtos_loja2.iloc[0]} unidades.")
print(f"Produto Menos Vendido: '{produto_menos_vendido_l2}' com {vendas_produtos_loja2.iloc[-1]} unidades.")

plt.figure()
top_10_loja2 = vendas_produtos_loja2.head(10)
sns.barplot(x=top_10_loja2.values, y=top_10_loja2.index, palette='plasma', hue=top_10_loja2.index, legend=False)
plt.title('Top 10 Produtos Mais Vendidos - Loja 2', fontsize=16)
plt.xlabel('Quantidade Vendida', fontsize=12)
plt.ylabel('Produto', fontsize=12)
plt.tight_layout()
plt.show()
print("\n\n")


# --- Loja 3 ---
print("Análise de Vendas - Loja 3")
print("="*50)
vendas_produtos_loja3 = loja3['Produto'].value_counts()
produto_mais_vendido_l3 = vendas_produtos_loja3.index[0]
produto_menos_vendido_l3 = vendas_produtos_loja3.index[-1]
print(f"Produto Mais Vendido: '{produto_mais_vendido_l3}' com {vendas_produtos_loja3.iloc[0]} unidades.")
print(f"Produto Menos Vendido: '{produto_menos_vendido_l3}' com {vendas_produtos_loja3.iloc[-1]} unidades.")

plt.figure()
top_10_loja3 = vendas_produtos_loja3.head(10)
sns.barplot(x=top_10_loja3.values, y=top_10_loja3.index, palette='magma', hue=top_10_loja3.index, legend=False)
plt.title('Top 10 Produtos Mais Vendidos - Loja 3', fontsize=16)
plt.xlabel('Quantidade Vendida', fontsize=12)
plt.ylabel('Produto', fontsize=12)
plt.tight_layout()
plt.show()
print("\n\n")


# --- Loja 4 ---
print("Análise de Vendas - Loja 4")
print("="*50)
vendas_produtos_loja4 = loja4['Produto'].value_counts()
produto_mais_vendido_l4 = vendas_produtos_loja4.index[0]
produto_menos_vendido_l4 = vendas_produtos_loja4.index[-1]
print(f"Produto Mais Vendido: '{produto_mais_vendido_l4}' com {vendas_produtos_loja4.iloc[0]} unidades.")
print(f"Produto Menos Vendido: '{produto_menos_vendido_l4}' com {vendas_produtos_loja4.iloc[-1]} unidades.")

plt.figure()
top_10_loja4 = vendas_produtos_loja4.head(10)
sns.barplot(x=top_10_loja4.values, y=top_10_loja4.index, palette='cividis', hue=top_10_loja4.index, legend=False)
plt.title('Top 10 Produtos Mais Vendidos - Loja 4', fontsize=16)
plt.xlabel('Quantidade Vendida', fontsize=12)
plt.ylabel('Produto', fontsize=12)
plt.tight_layout()
plt.show()
```

```python
# Adiciona uma coluna 'Loja' em cada DataFrame para identificação
loja['Loja'] = 'Loja 1'
loja2['Loja'] = 'Loja 2'
loja3['Loja'] = 'Loja 3'
loja4['Loja'] = 'Loja 4'

# Concatena todos em um único DataFrame
df_completo = pd.concat([loja, loja2, loja3, loja4])

print("Amostra do DataFrame unificado:")
df_completo.head()

# Calcular o faturamento total por loja
faturamento_por_loja = df_completo.groupby('Loja')['Preço'].sum().sort_values(ascending=False)

# Visualizar o faturamento
plt.figure(figsize=(10, 6))
sns.barplot(x=faturamento_por_loja.index, y=faturamento_por_loja.values, palette='rocket', hue=faturamento_por_loja.index, legend=False)
plt.title('Faturamento Total por Loja', fontsize=16)
plt.ylabel('Faturamento (R$)', fontsize=12)
plt.xlabel('Loja', fontsize=12)
plt.show()

# 1. Encontrar os 10 produtos mais vendidos no geral
top_10_produtos = df_completo['Produto'].value_counts().head(10).index

# 2. Filtrar o DataFrame para conter apenas esses produtos
df_top_produtos = df_completo[df_completo['Produto'].isin(top_10_produtos)]

# 3. Agrupar por Loja e Produto para contar as vendas
vendas_cruzadas = df_top_produtos.groupby(['Loja', 'Produto']).size().reset_index(name='Quantidade')

# 4. Pivotar a tabela para o formato do heatmap
heatmap_data = vendas_cruzadas.pivot_table(index='Produto', columns='Loja', values='Quantidade', fill_value=0)

# Reordenar as linhas para corresponder à ordem de mais vendidos
heatmap_data = heatmap_data.loc[top_10_produtos]

# 5. Desenhar o heatmap
plt.figure(figsize=(12, 10))
sns.heatmap(
    heatmap_data,
    annot=True,      # Mostra os números dentro de cada célula
    fmt='.0f',
    cmap='viridis',
    linewidths=.5
)
plt.title('Mapa de Vendas: Top 10 Produtos por Loja', fontsize=18)
plt.xlabel('Lojas', fontsize=12)
plt.ylabel('Produtos', fontsize=12)
plt.show()
```
